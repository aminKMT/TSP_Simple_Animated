```python
import itertools
import math

# Coordinates of cities
cities = {
    'A': (0, 0),
    'B': (1, 5),
    'C': (2, 3),
    'D': (4, 4),
    'E': (2, 6)
}

# Calculate Euclidean distance
def distance(a, b):
    x1, y1 = cities[a]
    x2, y2 = cities[b]
    return math.hypot(x2 - x1, y2 - y1)

# Brute-force TSP
def tsp_brute_force(cities):
    min_path = None
    min_dist = float('inf')
    for perm in itertools.permutations(cities.keys()):
        path = list(perm) + [perm[0]]  # return to start
        dist = sum(distance(path[i], path[i+1]) for i in range(len(path)-1))
        if dist < min_dist:
            min_dist = dist
            min_path = path
    return min_path, min_dist

best_path, best_dist = tsp_brute_force(cities)
print("Best Path:", best_path)
print("Total Distance:", best_dist)

```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

# --- Build coordinate lists from your best_path ----------------------------
x_coords = [cities[c][0] for c in best_path]
y_coords = [cities[c][1] for c in best_path]

# best_path already ends where it starts, so the loop is closed
# ---------------------------------------------------------------------------

# Interpolate between cities for smooth motion
path_x, path_y = [], []
steps_per_leg = 30          # make larger for an even smoother trace

for i in range(len(x_coords) - 1):
    xs = np.linspace(x_coords[i], x_coords[i+1], steps_per_leg)
    ys = np.linspace(y_coords[i], y_coords[i+1], steps_per_leg)
    path_x.extend(xs.tolist())
    path_y.extend(ys.tolist())

# --- Plot set‑up -----------------------------------------------------------
fig, ax = plt.subplots()
ax.set_aspect("equal", adjustable="box")
ax.set_xlim(min(x_coords) - 1, max(x_coords) + 1)
ax.set_ylim(min(y_coords) - 1, max(y_coords) + 1)
ax.set_title("Optimal TSP Route")

# Static city markers and labels
ax.plot(x_coords[:-1], y_coords[:-1], "ko")          # black dots
for city, (x, y) in cities.items():
    ax.text(x + 0.15, y + 0.15, city, fontsize=12)

# Line (route) and moving dot (salesman)
line, = ax.plot([], [], "b-", lw=2)
dot,  = ax.plot([], [], "ro", markersize=8)

def init():
    line.set_data([], [])
    dot.set_data([], [])
    return line, dot

def update(frame):
    line.set_data(path_x[:frame + 1], path_y[:frame + 1])
    # wrap single numbers in lists so they’re valid “sequences”
    dot.set_data([path_x[frame]], [path_y[frame]])
    return line, dot

ani = animation.FuncAnimation(
    fig, update,
    frames=len(path_x),
    init_func=init,
    interval=50,          # ms between frames
    blit=True,
    repeat=False
)

HTML(ani.to_jshtml())
```
